#Filtro dataset solo con fraudes

```python
fraud = credit_card_fraud[credit_card_fraud['is_fraud'] == 1]
```

**Gráfico en Barra Horizontal con colores pasteles para diferenciar las Categorías de los Fraudes**

```python
# Calculamos la frecuencia de cada categoría
category_counts = fraud['category'].value_counts()

# Ordenar las categorías por frecuencia
category_counts = category_counts.sort_values(ascending=True)

# Crear un gráfico de barras horizontales
plt.figure(figsize=(10, 6))  # Establecer el tamaño de la figura
sns.barplot(x=category_counts.values, y=category_counts.index,palette='pastel')

# Establecer etiquetas y título
plt.xlabel('Cantidades')
plt.ylabel('Categoría')
plt.title('Clasificación de Fraudes según sus Categorías')

# Mostrar el gráfico
plt.show()
```

Este gráfico muestra los diferentes formas de fraude y sus cantidades.





**Gráfico para género en barra horizontal**

```python
# Calculamos la frecuencia de fraudes por género
gender = fraud['gender'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=gender.index, y=gender.values,color='skyblue')

# Etiquetas y título
plt.xlabel('Género')
plt.ylabel('Cantidad de Fraudes')
plt.title('Cantidad de Fraudes por Género')

# Mostrar el gráfico
plt.show()
```

En esta ocasión queríamos saber que género era el que realizaba más fraudes.

**Gráfico de coordenadas de Latitud y Longitud**

```python
# Obtener las coordenadas de latitud y longitud
lat = fraud["lat"]
lon = fraud["long"]

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(lon, lat)

# Agregar etiquetas y título
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Ubicación de las transacciones de fraude")

# Mostrar el gráfico
plt.show()
```

Este código visualiza la ubicación de las transacciones de fraude con tarjeta de crédito en un gráfico de dispersión utilizando sus coordenadas geográficas.
